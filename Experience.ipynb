```python
def gen_prompts(topic):
    system_prompt = """
تو متخصص نوشتن پرامپت در موضوعات مختلف هستی. به تو موضوعی داده خواهد شد و وظیفه تو این است که دو نوع پرامپت مختلف در دو حالت مختلف بنویسی.

با توجه به موضوع داده شده، پرامپت اول یعنی پرامپت نقش، باید همانند یک system prompt نقش LLM را در ارتباط با پاسخگویی در حوزه موضوع مطرح شده مشخص کند.
همچنین پرامپت دوم یعنی پرامپت دستورالعمل، باید دستورالعمل یا همان instructions پاسخگویی به سوالات کاربران با توجه به سوال داده شده در ارتباط با موضوع مشخص شده را تعیین کند.

مثال:
موضوع: آشپزی جهانی و تکنیک‌های پخت
پرامپ نقش: شما یک متخصص آشپزی هستید که دستور العمل‌ها و تکنیک‌های آشپزی را مستند می‌کند.
مطالب شما باید دقیق و جامع باشد و شامل روش‌های پخت سنتی و مدرن باشد.
پرامپت دستوراعمل: دستور العمل‌ها و تکنیک‌های دقیق ایجاد کنید که شامل موارد زیر باشد:
- لیست مواد لازم با جایگزین‌های احتمالی
- دستورالعمل‌های گام به گام
- توضیحات تکنیک‌های مهم
- اشتباهات رایج که باید از آنها اجتناب کرد
- پیشنهادات نگهداری و سرو
- زمینه و تاریخچه فرهنگی

نکته بسیار مهم: هر دو پرامپت باید کاملاً وابسته به موضوع خواسته شده باشند و مثال داده شده فقط یک نمونه است و نباید دقیقاً اجزای آن کپی شود. همچنین هردو پرامپت باید به زبان فارسی باشند.

خروجی را در قالب JSON به فرمت زیر ارائه کن:
{
    "نقش": "<پرامپت نقش>",
    "دستورالعمل": "<پرامپت دستورالعمل>"
}
    """

    user_prompt = f"موضوع: {topic}"

    response = client.chat.completions.create(
        model="gpt-4o",
        messages=[
            {
                "role": "system",
                "content": system_prompt,
            },
            {
                "role": "user",
                "content": user_prompt,
            },
        ],
        response_format={ "type": "json_object" },
    )
    return response.choices[0].message.content
```

```python
import json

topic_prompts = []

for topic in topics.strip().split("\n"):
    topic = topic.split(".")[1].strip()
    prompts = json.loads(gen_prompts(topic))
    topic_prompts.append({"موضوع": topic, **prompts})

with open("prompts/topic_prompts.json", "w", encoding="utf-8") as f:
    json.dump(topic_prompts, f, ensure_ascii=False)
```

```python
from dotenv import load_dotenv

load_dotenv()

from promptwright import DataEngine, EngineArguments, TopicTree, TopicTreeArguments

def generate_dataset(topic, system_prompt, instructions):
    tree = TopicTree(
        args=TopicTreeArguments(
            root_prompt=topic,
            model_system_prompt=system_prompt,
            tree_degree=7,
            tree_depth=4,
            temperature=0.7,
            model_name="openai/gpt-4o",
        )
    )

    tree.build_tree()
    tree.save(f"topic_trees/{topic}_tree.jsonl")

    engine = DataEngine(
        args=EngineArguments(
            instructions=instructions,
            system_prompt=system_prompt,
            model_name="openai/gpt-4o",
            temperature=0.1,
            max_retries=3,
        )
    )

    dataset = engine.create_data(
        num_steps=200,
        batch_size=10,
        topic_tree=tree,
        sys_msg=False,
    )

    dataset.save(f"datasets/{topic}_dataset.jsonl")
```

```python
generate_dataset(topic_prompts[0]["موضوع"], topic_prompts[0]["نقش"], topic_prompts[0]["دستورالعمل"])
```
